```python
title_genres['genre_id']=title_genres['genres_split'].map(genre_map)
title_genres = title_genres.drop(columns='genres_split')
```

```python
#get max string length
max_str_len = title_basics['tconst'].fillna('').map(len).max()
```

```python
max_str_len
```

```python
from sqlalchemy.types import *
##Calculate max string lengths for basics columns
key_len = title_basics['tconst'].fillna('').map(len).max()
title_len=title_basics['primaryTitle'].fillna('').map(len).max()
##Create a schema dictionary using SQLalchemy datatype objects
basics_schema = {"tconst": String(key_len+1),
                "primaryTitle": Text(title_len+1),
                'startYear': Float(),
                'endYear':Float(),
                'runtimeMinutes':Integer()}
```

```python
#save to SQL with dtype and index=False
title_basics.to_sql('title_basics', engine, dtype=basics_schema, if_exists='replace', index=False)
```

```python
engine.execute('ALTER TABLE title_basics ADD PRIMARY KEY (`tconst`);')
```

```python
##Calculate max string lengths for ratings columns
key_len = title_ratings['tconst'].fillna('').map(len).max()
##Create a schema dictionary using SQLalchemy datatype objects
ratings_schema = {"tconst": String(key_len+1),
                'averageRating': Float(),
                'numVotes':Integer()}
```

```python
#save to SQL with dtype and index=False
title_ratings.to_sql('title_ratings', engine, dtype=ratings_schema, if_exists='replace', index=False)
```

```python
engine.execute('ALTER TABLE title_ratings ADD PRIMARY KEY (`tconst`);')
```

```python
##Calculate max string lengths for tmdb columns
key_len = tmdb_merged['tconst'].fillna('').map(len).max()
title_len=tmdb_merged['certification'].fillna('').map(len).max()
##Create a schema dictionary using SQLalchemy datatype objects
tmdb_schema = {"tconst": String(key_len+1),
                'revenue': Float(),
                'budget':Float(),
                'certification':Text(title_len+1)}
```

```python
#save to SQL with dtype and index=False
tmdb_merged.to_sql('tmdb_merged', engine, dtype=tmdb_schema, if_exists='replace', index=False)
```

```python
engine.execute('ALTER TABLE tmdb_merged ADD PRIMARY KEY (`tconst`);')
```
