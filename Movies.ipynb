```python
import pandas as pd
import numpy as np
```

```python
basics = pd.read_csv("MovieData/title_basics.csv.gz", low_memory=False)
basics.head()
```

```python
basics.replace({'\\N':np.nan})
```

```python
basics = basics.dropna(subset=["runtimeMinutes"])
```

```python
# Confirming no more null sale prices
basics['runtimeMinutes'].isna().sum()
```

```python
basics = basics.dropna(subset=['genres'])
```

```python
basics['genres'].isna().sum()
```

```python
basics = basics.dropna(subset=['startYear'])
```

```python
basics['startYear'].isna().sum()
```

```python
basics['startYear'].unique()
```

```python
basics = basics.loc[basics['startYear'] != '\\N']
```

```python
basics['startYear'].unique()
```

```python
basics['startYear'] = basics['startYear'].astype(float)
print(basics.dtypes)
```

```python
basics = basics[(basics['titleType']=='movie') & (basics['startYear'] > 1999)]
basics.head()
```

```python
is_documentary = basics['genres'].str.contains('documentary', case = False)
basics = basics[~is_documentary]
```

```python
akas = pd.read_csv("MovieData/title_akas.csv.gz", low_memory=False)
akas.head()
```

```python
akas.replace({'\\N':np.nan})
```

```python
akas = akas[(akas['region']=='US')]
```

```python
keepers = basics['tconst'].isin(akas['titleId'])
keepers
```

```python
basics = basics[keepers]
basics
```

```python
ratings = pd.read_csv("MovieData/title_ratings.csv.gz", low_memory=False)
ratings.head()
```

```python
ratings.replace({'\\N':np.nan})
```

```python
keepers2 = ratings['tconst'].isin(akas['titleId'])
keepers2
```

```python
ratings = ratings[keepers2]
ratings
```
