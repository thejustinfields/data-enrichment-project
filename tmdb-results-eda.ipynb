```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies_2000 = pd.read_csv('Data/final_tmdb_data2000.csv.gz')
movies_2001 = pd.read_csv('Data/final_tmdb_data2001.csv.gz')
```

```python
movies_2000.head()
```

```python
movies_2001.head()
```

```python
merged_df = pd.concat([movies_2000, movies_2001], axis = 0)
```

```python
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df = merged_df.dropna(subset=["budget"])
```

```python
merged_df = merged_df.dropna(subset=["revenue"])
```

```python
merged_df['budget'].isna().sum()
```

```python
merged_df['revenue'].isna().sum()
```

```python
merged_df['revenue'].describe()
```

```python
merged_df = merged_df.loc[merged_df['revenue']>0]
```

```python
merged_df=merged_df.loc[merged_df['budget']>0]
```

```python
merged_df['revenue'].describe()
```

```python
merged_df['budget'].describe()
```

```python
#explore Budget
ax = merged_df['budget'].hist(bins = 30, edgecolor = 'black')
ax.tick_params(axis='x', rotation = 45)
ax.ticklabel_format(style='plain')
ax.set_title('Movie Budget Totals')
ax.set_xlabel('Budget Totals')
ax.set_ylabel('Number of Movies');
```

- There are around 150 movies with budgets less than \\$20,000,000.

- There are more than 90 movies with budgets between \\$20,000,000 and \\$40,000,000.

- There are more than 30 movies with budgets between \\$40,000,000 and \\$60,000,000.

- There are more than 30 movies with budgets between \\$60,000,000 and $80,000,00

- There are more than 20 movies with budgets between \\$80,000,000 and \\$100,000,000

- There are more than 20 movies with budgets between \\$100,000,000 and \\$140,000,000


```python
#Explore Certification categories
ax = merged_df['certification'].hist(edgecolor = 'black')
ax.tick_params(axis='x', rotation = 45)
ax.set_title('Movie Totals by Certification')
ax.set_xlabel('Certification Categories')
ax.set_ylabel('Number of Movies');
```

- There are over 140 movies that are rated R
- There are less than 120 movies that are rated PG-13
- There are around 30 movies that are rated PG
- There are less than 20 movies rated G
- There are less than 10 movies that are Not Rated.

```python
# Plot the average of each as a bar
# Using seaborn to plot the group means
ax = sns.barplot(data=merged_df, x='certification', y="budget").set(title='Average Budget by Certification')
```

- The average budget for Rated R movies is less than \\$30,000,000.
- The average revenue for Rated PG-13 movies is less than \\$50,000,000.
- The average revenue for Rated PG movies is more than \\$50,000,000.
- The average revenue for Rated G movies is more than \\$40,000,000.
- The average revenue for Not Rated movies is less than \\$20,000,000.
